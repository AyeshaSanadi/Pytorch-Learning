Reduction operation

```python
data2 = torch.randint(low=9, high=24, size=(2,3), dtype=torch.float32)
data2
```

```python
# sum
print(torch.sum(data2))

# row sum
print(torch.sum(data2, dim=1))

# column sum
print(torch.sum(data2, dim=0))
```

```python
# mean
print(torch.mean(data2))
# mean on column
print(torch.mean(data2, dim=0))
# mean on row
print(torch.mean(data2, dim=1))
```

```python
# min
print(torch.min(data2))
# max
print(torch.max(data2))
# product
print(torch.prod(data2))
# standard deviation
print(torch.std(data2))
# variance
print(torch.var(data2))
# argmax -> position of max argument
print(torch.argmax(data2))
# argmin -> position of min argument
print(torch.argmin(data2))
```

Comparison Operation on each element

```python
t1 = torch.randint(low=1, high=100, size=(3,3), dtype=torch.float32)
t2 = torch.randint(low=1, high=100, size=(3,3), dtype=torch.float32)
print(t1)
print(t2)
```

```python
print(t1 > t2)
print(t1 >= t2)
print(t1 < t2)
print(t1 <= t2)
print(t1 == t2)
print(t1 != t2)
```

Tensor operation on GPU

```python
import torch
torch.cuda.is_available()
```

```python
device = torch.device('cuda')
device
```

```python
# creating tensor
torch.tensor([1,2,3,4,5], device=device)
```

```python
t1 = torch.rand(2,3)
t1
```

```python
# running existing tensors inside gpu
t1.to(device)
```

```python
import time

size = 10000

matrix1 = torch.randn(size,size)
matrix2 = torch.randn(size,size)

# measure time on cpu
s = time.time()
result_cpu = torch.matmul(matrix1, matrix2)
e = time.time()
print(f"Total time required: {e-s} seconds")

# measure time on gpu
s = time.time()
matrix1 = matrix1.to('cuda')
matrix2 = matrix2.to('cuda')
result_cpu = torch.matmul(matrix1, matrix2)
e = time.time()
print(f"Total time required: {e-s} seconds")
```
